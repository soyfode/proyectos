# FIFA 19 dataset de jugadores completo

```python
import pandas as pd
```

```python
data=pd.read_csv("data.csv")
```

```python
data.drop(['ID','Unnamed: 0','Photo','Club Logo','Flag'],axis=1).head()
```

## Distribución normal estándar

```python
mean=data.Age.mean()
std=data.Age.std()
```

```python
age=[]
for i in data.Age:
    age.append((i-mean)/std)
```

```python
sum(age)
```
