# Paris

```python
# library
import pandas as pd
import matplotlib.pyplot as plt
import contextily
```

```python
# data
restaurants = pd.read_csv("data/Paris/paris_restaurants.csv")
```

```python
#Visualization
restaurants.head()
```

```python
# figura de todos los restaurants
fig, ax = plt.subplots()
ax.plot(restaurants.x,restaurants.y,'o',markersize=1)
contextily.add_basemap(ax)
plt.show()
```
