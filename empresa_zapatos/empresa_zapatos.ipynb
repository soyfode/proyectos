## ¿Cuál es el número de zapatos que probablemente se vendan, basados en datos históricos?

#### Estratégia: Hallar el intervalo de confianza del 95%
1. Últimos doce meses de ventas.
2. Solo para zapatos de hombre (ya que el problema es idéntico)
3. Solo para los EEUU (Ya que el problema es identico)

```python
mujer=pd.pivot_table(data[data['Género']=='Mujer'].loc[:,['Pais','Talla (EE.UU.)']],index='Talla (EE.UU.)',columns='Pais',aggfunc=len)
mujer
```

```python
hombre = pd.pivot_table(data[data['Género']=='Hombre'].loc[:,['Pais','Talla (EE.UU.)']],index='Talla (EE.UU.)',columns='Pais',aggfunc=len)
hombre
```

#### Calculamos la cantidad vendida por cada mes para tallas de hombre

```python
talla_mes=pd.pivot_table(data[data['Género']=='Hombre'].loc[:,['Talla (EE.UU.)','Month']],index='Talla (EE.UU.)',columns='Month', aggfunc=len)
talla_mes
```

#### Calculamos las medias de cada fila

```python
media=talla_mes.mean(axis=1)
```

#### La varianza es desconocida, luego buscamos la estadistica "t"

#### Calculamos el error estandar

```python
error=talla_mes.std(axis=1)
```

#### Calculamos los intervalos de confianza media

```python
sup=talla_mes.mean(axis=1)+(2.18*talla_mes.std(axis=1))
inf=talla_mes.mean(axis=1)-(2.18*talla_mes.std(axis=1))
```

#### Creamos una tabla

```python
pd.DataFrame({'Media':media,'Error_standar':error,'Limite_inf':inf,'Limite_sup':sup,'Pares_de_zapatos':round(sup)})
```

- El límite superior nos muestra el número máximo de pares requeridosluego de redondear los mismos.
